## Introduction

This notebook provides a simple illustration of how to run regression analysis in Python. It assumes that users are already familiar with regression statistical concepts. 

We will use the the Student Performance Dataset from Kaggle. The dataset consists of 10,000 student records, with each record containing information about various predictors and a performance index.

**Variables:**

**Hours Studied:** The total number of hours spent studying by each student.

**Previous Scores:** The scores obtained by students in previous tests.

**Extracurricular Activities:** Whether the student participates in extracurricular activities (Yes or No).

**Sleep Hours:** The average number of hours of sleep the student had per day.

**Sample Question Papers Practiced:** The number of sample question papers the student practiced.

**Performance Index:** A measure of the overall performance of each student. The performance index represents the student's academic performance and has been rounded to the nearest integer. The index ranges from 10 to 100, with higher values indicating better performance.

## Dataset Overview and Variable Description

Here we will read in the data and take a peek. 

```python
# Load the dataset
df = pd.read_csv('data/ParisHousing.csv')

# Display the first few rows of the dataframe
print(df.head())
```

```python
# Display summary stats
print(df.describe(include='all'))
    
```

## Question 1: What is the relationship between sleep hours and student performance? 

As an example, let's look at the relationship between sleep hours and student performance? First, we can visualize the relationship to help assess the linearity and identify potential outliers.
    

```python
sns.scatterplot(x='garage', y='price', data=df)
plt.xlabel('squareMeters')
plt.ylabel('price')
plt.title('House Size vs. House Price')
plt.show()
    
```


Fitting a linear regression model to quantify the relationship between median income and median house value.
    

```python

X = sm.add_constant(df['MedInc'])  # Add a constant term for the intercept
Y = df['MedianHouseValue']

model_1 = sm.OLS(Y, X).fit()
print(model_1.summary())
    
```
